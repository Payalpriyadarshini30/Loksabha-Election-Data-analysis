```python
def calculate_margin(df):
    margin_list = []
    for _, group in df.groupby(['state', 'pc_name']):
        group = group.sort_values(by='total_votes', ascending=False)
        winner = group.iloc[0]
        if len(group) > 1:
            runner_up = group.iloc[1]
            margin = winner['total_votes'] - runner_up['total_votes']
        else:
            margin = winner['total_votes']
        margin_list.append([winner['state'], winner['pc_name'], winner['candidate'], margin])
    margin_df = pd.DataFrame(margin_list, columns=['state', 'pc_name', 'candidate', 'margin'])
    return margin_df
margins_2014 = calculate_margin(df)
margins_2019 = calculate_margin(df2)
```
